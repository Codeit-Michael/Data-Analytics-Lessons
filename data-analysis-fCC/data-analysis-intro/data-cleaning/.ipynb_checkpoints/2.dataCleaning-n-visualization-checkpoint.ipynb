# Data Cleaning and Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Matplolib APIs
- Global API - commonly used, inspired by MATLAB
- OOP Approach

## Global vs OOP

### Global API

```python
x = np.arange(-10,11)
```

```python
# global API example
plt.figure(figsize=(12, 6)) # starting a graph
plt.title("My Plot")

plt.subplot(1,2,1) #rows, columns, ppanel selected
plt.plot(x,x**2) # plotting bell shape
plt.plot([0,0], [-10,100]) # [nth x points], [nth y points]
plt.legend(['X^2','Vertical Line']) # legends
plt.xlabel("X")
plt.ylabel("X Squared")

plt.subplot(1,2,2)
plt.plot(x, -x**2)
plt.plot([-10,10],[-100,0])
plt.legend(["X^2", "Diagonal Line"])
plt.xlabel("x")
plt.ylabel("X Squared")
```

The problem with **Global API** is it was created line-by-line. If you're working with bigger data and have to debug, you have to remember what each line of code we have above are doing (unless you put a cooment for each line). 

### OOP Interface

```python
plot_objects = plt.subplots(nrows=2, ncols=2, figsize=(14, 6))

fig, ((ax1, ax2), (ax3, ax4)) = plot_objects

plot_objects
```

Unlike in **OOP Approach**, you can easily access different rows. You can also modify it in different orders you want, unlike in Global Approach where you need to follow it line-by-line.

```python
ax4.plot(np.random.randn(50), c='yellow')
ax1.plot(np.random.randn(50), c='red', linestyle='--')
ax2.plot(np.random.randn(50), c='green', linestyle=':')
ax3.plot(np.random.randn(50), c='blue', marker='o', linewidth=3.0)

fig
```
